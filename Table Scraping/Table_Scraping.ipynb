# Webscrap for stock data performance 

```python
import pandas as pd
import numpy as np
import yfinance as yf

data = pd.read_csv("../data/ticker.csv",index_col=0)#read from ticker.csv
data = data[data['Ticker']!="No result"] #filter out companies that return no result for ticker symbol
tickers=[]

#loop starts here
for i in range(len(data.index)):
    tickerdata = yf.Ticker(data['Ticker'].values[i])#not case sensitive
    history = tickerdata.history(start="2019-06-01", end="2020-05-01") #335 days
    trial = pd.DataFrame(history)
    if(len(trial.index)==0):
        continue #if no data, skip this ticker
    else:
        total=0
        for j in range(len(trial.index)):
            avg = (trial['Close'].values[j]-trial['Open'].values[j])/trial['Open'].values[j]#gets the difference in percentage
            total+=avg #add to total
        percentage=(total/len(trial.index))*100 #grab percentage change
        round(percentage,2)
        tickers.append([data['Ticker'].values[i],round(percentage,2)]) #append to tickers list
```

```python
tickers_df = pd.DataFrame(tickers,columns=["Symbol","Average Percentage Growth"])
tickers_df.head(20)
#Write in random to_csv for average growth
```
